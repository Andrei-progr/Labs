### Сингулярное разложение


```python
import math
class SVD():

    def __init__(self, matrix):
        self.matrix = matrix
        self.size = min(matrix.shape[0], matrix.shape[1])
        self.V = np.eye(self.size)

    # Точность
    def ND(self,A):
        res = 0
        for i in range(1,self.size):
            for k in range(i):
                res += A[i,k]**2
        return math.sqrt(res)

    # Матрица Гивенса
    def G(self,j,k,c,s):
        a = np.eye(self.size)
        a[j,j] = c #math.cos(theta)
        a[k,j] = s #math.sin(theta)
        a[j,k] = -s #-math.sin(theta)
        a[k,k] = c #math.cos(theta)
        return a

    def Gt(self,j,k,c,s):
        a = np.eye(self.size)
        a[j,j] = c #math.cos(theta)
        a[j,k] = s #math.sin(theta)
        a[k,j] = -s #-math.sin(theta)
        a[k,k] = c #math.cos(theta)
        return a

    def sign(self, x):
        if x >= 0:
            return 1
        else:
            return -1

    # Метод Якоби вычисления сч и св для симметрической матрицы
    def Jakobi(self,A):
        lenght = A.shape[0]
        for i in range(1,lenght):
            for k in range(i):
                if abs(A[i,k]) > 10**(-8):
                    tau = (A[i,i] - A[k,k])/(2*A[i,k])
                    if abs(tau) < 10**(-14):
                        theta = math.pi/4
                        c = math.cos(theta)
                        s = math.sin(theta)
                    else:
                        t = 1/(tau + self.sign(tau)*math.sqrt(tau**2 + 1))
                        c = 1/math.sqrt(t**2 + 1)
                        s = t*c
                    A = np.dot(self.Gt(i, k, c, s), A)
                    A = np.dot(A, self.G(i, k, c, s))
                    self.V = np.dot(self.V, self.G(i,k,c,s))
        return A

    # Возвращает список сингулярных чисел и ортогональную матрицу правых собственных векторов
    def pca(self,epsilon):
        B = np.dot(self.matrix.transpose(), self.matrix)
        B = B / (max(self.matrix.shape[0], self.matrix.shape[1]) - 1)
        while self.ND(B) > epsilon:
            B = self.Jakobi(B)

        sing_val = []
        for i in range(B.shape[0]):
            sing_val.append(math.sqrt(abs(B[i,i])))
        return sing_val, self.V
```

```python
H = data2.copy()

svd = SVD(H)
eps = 10**(-8)
singular, V = svd.pca(eps)
singular
```
